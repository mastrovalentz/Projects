## Ex 2.0

Make plot of the principal discrete pdfs and cdfs

```r
x <- 0:5
prob <- c(0.1, 0.5, 0.9)
t <- 5
angs=c(-45, 0, 45)

cols <- c("red", "blue", "green", "yellow", "chartreuse4", "cyan2", "darkorchid2")
for (i in 1:length(prob)){
    ifelse(i==1,
        barplot(dbinom(x,t, prob[i]), names=x, legend=TRUE, col=cols[i], space=0,angle=angs[i], density=30, 
                xlab='x', ylab="f(x)", main="Binomial PDFs comparison"),
        barplot(dbinom(x,t, prob[i]), names=x, legend=TRUE, col=cols[i], space=0,angle=angs[i], density=30, add=TRUE))
}
legend("top",c(paste("P=", prob)), fill=cols[1:length(prob)])
```

## EX 2.1

The following tables reports a set of values (x), with the corresponding
probabilities given by the two methods (p1 and p2)

```r
x  <- c(15.58 , 15.9, 16  , 16.1, 16.2)
p1 <- c(0.15  , 0.21, 0.35, 0.15, 0.14)
p2 <- c(0.14  , 0.05, 0.64, 0.08, 0.09)
```

### 2.1.1
Evaluate the expected values, E[X], and the variance, Var(X), for both methods:

The definition for the expected value and variance are:

$E[X]=\sum_i x_i p(x_i)$

$var(X)=E[X^2]-E[X]^2$

We can directly plug the numbers in those definition and we obtain:

```r
e_xp1 <- sum(x*p1) #expected value for first probabilities
e_xp2 <- sum(x*p2)

v_xp1 <- sum(x*x*p1)-e_xp1^2
v_xp2 <- sum(x*x*p2)-e_xp2^2

print("First distribution: ")
print(paste("E[x]=", e_xp1, " var(x)=",v_xp1))
print("Second distribution: ")
print(paste("E[x]=", e_xp2, " var(x)=",v_xp2))
```

## EX 2.2

The waiting time, in minutes, at the doctors is about 30 minutes, and the distribution follows an exponential pdf with rate 1/30.

### 2.2.A
Simulate the waiting time for 50 people at the doctors office and plot the relative
histogram

I can use the sampling functionality for the exponantial distribution to obtain the simulated data

```r
rate <- 1/30
n <- 50
wait_times <- rexp(n, rate) #data sampling fromn exponential

hist(wait_times, col="red", breaks=15,
     xlab="wait time (min)", ylab="Counts", main="Waiting time histogram")
```

### 2.2.B
What is the probability that a person will wait for less than 10 minutes ?

We can compute this probability using the CDF exponantial function of the exponential since the CDF represents the integral up to a certain value.

$p(x\leq10) =CDF(x=10)$

```r
prob_under10 <- pexp (10, rate)
print(paste("Prob of waiting less than 10 min:" ,prob_under10))
```

This can be easily verified by sampling a big number of data from the PDF and counting the fraction of points below our threshold

```r
#verify
sample <- rexp(50000, rate) 
n_under10 <- length(sample[sample<10]) #number of events under 10 min
 #print the fraction which corresponds to the probability
print(paste("Prob of waiting len than 10 min:" ,n_under10/50000))
```

### 2.2.C
Evaluate the average waiting time from the simulated data and compare it with
the expected value (calculated from theory and by manipulating the probability
distributions using R)

The theoretical value is:

$E[X] =\int_{R+} \lambda e^{-\lambda x}$

$=... \text{per parti} ...=\int_0^{+\infty} e^{-\lambda x}=\frac{1}{\lambda} $

In this computation $\tau$ is the mean wiating time so the expected mean value from the theory is 30 min

To evaluate the mean wait time we can also compute the mean of the values that we simulated or manupulating the PDF by generating uniform values in an interval and computing the mean value weighting their probabilities:

```r
mean_wait_th  <- 1/rate
mean_wait_sim <- mean(wait_times)

x <- seq(0, 1000, 1) #i put 1000 even thou theoretically i should sum up to infinity 
                    # but the PDF at 1000 and more can be considered 0 in this case
mean_wait_pdf <- sum(x* dexp(x, rate))

print(paste("Mean value by theory:", mean_wait_th ))
print(paste("Mean value by sample:", mean_wait_sim))
print(paste("Mean value by PDFS:  ", mean_wait_pdf))
```

As we can see the three result are compatible but the value obtained from the simulated data is a little different, this may be due to the low number of simulated events, in fact as we can see below by simulating more events the result is almost equal to the theoretical value

```r
sample <- rexp(1000000, rate) 
mean_sampled <- mean(sample)
print(paste("Mean value by more sampling: ",mean_sampled))
```

### 2.2.D
What is the probability for waiting more than one hour before being received ?

As we did in point 2.2.B we can compute this probability using the CDF of the exponential distribution and verify the result by sampling a big number of event and computing the fraction that respects our condition

```r
prob_over1h_th <- 1-pexp(60, rate) #theoretical vale
#verify by sampling
sample <- rexp(100000, rate) 
n_over1h <- length(sample[sample>60])

print(paste("Mean value by theory:", prob_over1h_th ))
print(paste("Mean value by sample:", n_over1h/100000))
```

## EX 2.3

Lets suppose that on a book, on average, there is one typo error every three pages.

### 2.3.1
If the number of errors follows a Poisson distribution, plot the pdf and cdf

```r
options(repr.plot.width=16, repr.plot.height=8)
area <- matrix (c(1,2), nrow=1, ncol=2, byrow=TRUE)
layout(area)

lambda <- 1/3
x<- seq(-1,6, 1) #the -1 is just to make the plot better estetichally, it's probability is 0
plot(x, dpois(x, lambda), xlim=c(0,6), ylim=c(0,0.8), xlab='Number of errors', ylab="PDF", main="Poisson PDF lambda=1/3", pch=23, cex=1)
lines(x, dpois(x, lambda), xlim=c(0,6), ylim=c(0,0.8), xlab='x', ylab="f(x)", main="Poisson PDF lambda=1/3",lwd=2.5, type="s", col="red")

plot(x, ppois(x, lambda), xlim=c(0,6), ylim=c(0,1), xlab='Number of errors', ylab="CDF", main="Poisson CDF lambda=1/3", pch=23, cex=1)
lines(x, ppois(x, lambda),  xlab='x', ylab="f(x)", main="Poisson CDF lambda=1/3", type="s", lwd=2,col="red")
```

### 2.3.2
Calculate the probability that there is at least one error on a specific page of the
book

The probability of at least one error can be computed as 1 minus the probability of not having error, in this case it can be computed directly from the distribution PDF

```r
prob_more1 <- 1-dpois(0, 1/3)
print(paste("Probability of not having errors: ", prob_more1))
```

## Ex 2.4

We randomly draw cards from a deck of 52 cards, with replacement, until one ace is
drawn. Calculate the probability that at least 10 draws are needed

Starting from the probability of drawing an ace i can compute this probability in some different ways:

- using the binomial distibution and asking to have 0 success in drawing an ace in the first 9 draws
- using the negative binomial and asking 0 failures in not drawing an ace in the first 9 draws
- using the geometric distribution cumulative density function and subtrancting it's value to 1
- it the simplest way by doing this reasoning: to draw at least 10 cards i need that the first 9 draws are not aces so the probability of drawing 10 cards at least will be the probability of not drawing an ace to the 9th power

```r
p <- 4/52
prob_bin <- dbinom(0,9, p)
prob_nbin <-  dnbinom(0, 9, 1-p)
prob_exp <-  (1-p)^9
prob_geom <- 1 - sum(pgeom(8, p ))

print(paste("Probability by binomial:    ",prob_bin))
print(paste("Probability by neg binomial:",prob_nbin))
print(paste("Probability by elevation:   ",prob_exp))
print(paste("Probability by geometric:   ",prob_geom))
```

As we can see all the methods leads to the same result but we can also verify the goodness of our results by simulating some events: in the code below i generated a random matrix with values from 1 to 52, each row has 9 columns representing the drawn cards.

The number whose rest quotient by 13 (1,14,27, 40) represent the aces, so to know the fraction on events in which less than 10 cards are drawn i just have to calculate the number of rows where those numbers appears, the probability of drawing at least 10 cards is then 1 minus this fraction

```r
nr <- 1000000
nc <- 9

m <- matrix(sample.int(52, size = nr*nc, replace = TRUE), nrow = nr, ncol = nc)
r <- rowSums(m%%13==1, na.rm=TRUE)#>0)*1L

#print(length(r[r>0]))
print(paste("Probability by simulation:   ",1-length(r[r>0])/nr))
```

As we can see the results are perfectly compatible.

### 2.5.4

Plot a distribution of the time the mayor is in charge. The starting date is in column
data elezione. Since elections happen every 5 years, how many of them are going
to complete their mandate this year ? And how many in 2021 ?

```r
today <- as.Date(Sys.Date(), format='%d/%m/%Y')
charge_distr <- as.numeric(difftime( today, 
               as.Date(dataTibble$data_elezione, format="%d/%m/%Y") ))%/%365
#print(ages_distr)

barplot(table(as.integer(charge_distr)), col='blue')
```

To compute how many mayors will end their mandate this year we can create a vector containing the difference in years or the time passed from their election to the end of the year. The number of people ending their mandate this year will then be the ones whose time diffrence is bigger than 5 (duration of a mandate). 

The same thing can be done to count how many will end their mandate next year by using the end of next years and adding the constrain of the time difference being not only bigger tahn 5 but also less than 6 to avoid double counting the ones ending the mandate this year

```r
dates <- as.Date(dataTibble$data_elezione, format="%d/%m/%Y")

until20 <- difftime(as.Date("31/12/2020",  format="%d/%m/%Y" ), dates, unit='days')/365
until21 <- difftime(as.Date("31/12/2021",  format="%d/%m/%Y" ), dates, unit='days')/365

print(paste("Mandates ending in 2020: ", length(until20[until20>5 ] )))
print(paste("Mandates ending in 2021: ", length(until21[until21>5 & until21 <=6] )))
```
