## Ex 1.1

Given the following data create a data frame containing the information about lakes build a data frame

```r
Loch <- c("Ness", "Lomond" ,"Morar","Tay","Awe", "Maree","Ericht",
        "Lochy", "Rannoch","Shiel",  "Katrine","Arkaig","Shin")
Volume <- c(7.45,2.6,2.3,1.6,1.2,1.09,1.08,1.07,
            0.97,0.79,0.77,0.75,0.35)
Area <- c(56,71,27,26.4,39,28.6,18.6,16,
         19,19.5,12.4,16,22.5)
Lenght <- c(39,36,18.8,23,41,20,23,16,
            15.7,28,12.9,19.3,27.8)   
MaxDepth <- c(230,190,310,150,94,114,156,
             162,134,128,151,109,49)
MeanDepth <- c(132,37,87,60.6,32,38,57.6,
            70,51,40,43.4,46.5,15.5 )

data <- data.frame (Loch, Volume, Area, Lenght, MaxDepth, MeanDepth)
print(data)
```

### 1.2.1
Create a table grouping the data by gender and by sport; produce a barplot with the table adding
a legend

```r
#first of all i need a vector with the sports
sports <- unique(dataTibble$sport)
#i create the table
SportTable <- matrix(0, nrow = 2, ncol = length(sports))
colnames(SportTable) <- sports
rownames(SportTable) <- c('Male', 'Female')
#fill the table by measuring the lenght of the tibble(columnwise) after filtering by gender and sport
for(i in 1:length(sports)){
    SportTable[1,i]=length(filter(dataTibble,sex=='m', sport==sports[i])$sport)
    SportTable[2,i]=length(filter(dataTibble,sex=='f', sport==sports[i])$sport)
}
print(SportTable)


```

```r
#the table is done so i can make a barplot from it
barplot(SportTable, beside=TRUE, 
        col=c('cyan', 'red'),
        legend.text=TRUE, 
        main="Number of athlete per sport, gender comparison",
        xlab="Sports",
        ylab="NUmber of athletes",
        ylim=c(0,25),
        las=2)
```

```r
#this is an alternative way of getting the table but i preferred the one above since 
#this is more difficult to be plotted due to some sport being praticated by only males or females
GenderSportTibble <- tibble (gender = dataTibble$sex, sport=dataTibble$sport)
table <- GenderSportTibble %>% count(sport,gender)
print(table)
```

### 1.2.3
Produce boxplots of the main blood variables (’red blood cell counts’, ’white blood cell counts’,
’hematocrit’ and ’hemaglobin concentration’), for different kind of sports

```r
library(repr)
options(repr.plot.width=12, repr.plot.height=12)
area <- matrix (c(1,2,3,4), nrow=2, ncol=2, byrow=TRUE)
layout(area)
boxplot(rcc~sport, data= filter(dataTibble, ),las=2, col ='green', main="Red blood cell count comparison per sport")
boxplot(wcc~sport, data= filter(dataTibble, ),las=2, col ='green', main="White blood cell count comparison per sport")
boxplot( hc~sport, data= filter(dataTibble, ),las=2, col ='green', main="Hematocrit comparison per sport")
boxplot( hg~sport, data= filter(dataTibble, ),las=2, col ='green', main="Hemaglobin concentration comparison per sport")
```

## Ex 1.3

Download the latest update on the COVID-19 Virus infection fom the European Centers for Disease
Control

```r
library(lubridate)
library(readxl)
library(curl)
url <- "https://www.ecdc.europa.eu/sites/default/files/documents/"
fname <- "COVID-19-geographic-disbtribution-worldwide-"
date <- lubridate::today()-1
ext = ".xlsx"
target <- paste(url,fname,date,ext, sep="" )
message ( "target: " , target)
tmp_file <- tempfile( "data" , "/tmp" , fileext = ext )
tmp <- curl::curl_download (target , destfile=tmp_file )
covid <- readxl::read_xlsx ( tmp_file )
print(covid)
```

Examine the loaded tibble structure

```r
str(covid)
```

### 1.3.2
Create a sub-tibble containing only the last day  and
produce a table with all the countries with number of deaths or number of new cases greater than
200

```r
lastDay <- covid$dateRep[1] #this is the last day
lastDayCovid <- filter(covid , dateRep==lastDay) #apply a filter
print(lastDayCovid)
```

```r
over200Covid <- filter(lastDayCovid, cases>200 | deaths>200 ) #filter the tibble
table <- data.frame(over200Covid$countriesAndTerritories, over200Covid$cases, over200Covid$deaths) #produce the table
table <- column_to_rownames(table, names(table[1])) #make the states become rownames
names(table) <- c("Cases", "Deaths") #set column names
print(table)
```

### 1.3.3
Select the top 10 countries, in terms of cases, and plot the total number of cases as a function of
time. Plot the total number of deaths as a function of time. In order to compare the different
curves, normalize the first date-time plot to the same $t_0$ value.

```r
top10Country <- row.names(table[order(-table$Cases),])[1:10]
print(top10Country)
```

In the following plot the number of cases and deaths are normalized starting from the day of the first registered case in the top 10 countries, the curves are proposed in two ways: in normal scale and in logaritmic scale which is easier to confront.

All the curves start from the 31/12 the day of the first registered case, if wanted is possible to show only more recent data by adding a second filter in the plot.

The advantage of the logaritmic scale (aside from the easier visualization) is that each curve is only plotted starting from the day of the first case.

```r
options(warn=-1) #deactivate logaritmic warning

library("RColorBrewer")
options(repr.plot.width=16, repr.plot.height=16)
area <- matrix (c(1,2,3,4), nrow=2, ncol=2, byrow=TRUE)
layout(area)

ymax=sum(filter(covid, countriesAndTerritories==top10Country[1])$cases)
for (i in 1:10){
    plot(cumsum(cases)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i] ), 
         xlab='', ylab='Total Cases', ylim=c(0, ymax), 
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n', cex=1.5)
    lines(cumsum(cases)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i] ), xlab='',
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n',yaxt='n', lwd=3)
    par(new=TRUE)
}
legend("topleft", legend=top10Country, col=brewer.pal(n = 10, name = "Spectral"),  lty=1, lwd=4, bty='n', y.intersp=0.5)
axis.POSIXct(1, at=seq(min(covid$dateRep), max(covid$dateRep), by="5 days"), format="%m-%d", las=2)
axis(2, at=seq(0,ymax, 100000), label=seq(0,ymax, 100000) )

ymax=sum(filter(covid, countriesAndTerritories==top10Country[1])$deaths)
for (i in 1:10){
    plot(cumsum(deaths)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i] ), 
         xlab='', ylab='Total Deaths',ylim=c(0, ymax),
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n', yaxt='n', cex=1.5)
    lines(cumsum(deaths)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i] ), xlab='',
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n',yaxt='n', lwd=3)
    par(new=TRUE)
}
legend("topleft", legend=top10Country, col=brewer.pal(n = 10, name = "Spectral"),  lty=1, lwd=4, bty='n',y.intersp=0.5)
axis.POSIXct(1, at=seq(min(covid$dateRep), max(covid$dateRep), by="5 days"), format="%m-%d", las=2)
axis(2, at=seq(0,ymax, 2000), label=seq(0,ymax, 2000) )




ymax=sum(filter(covid, countriesAndTerritories==top10Country[1])$cases)
for (i in 1:10){
    plot(cumsum(cases)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i] ), 
         xlab='', ylab='Total Cases', ylim=c(1, ymax), log='y',
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n', cex=1.5)
    lines(cumsum(cases)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i] ), xlab='',
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n',yaxt='n', lwd=3)
    par(new=TRUE)
}
legend("topleft", legend=top10Country, col=brewer.pal(n = 10, name = "Spectral"),  lty=1, lwd=4, bty='n', y.intersp=0.5)
axis.POSIXct(1, at=seq(min(covid$dateRep), max(covid$dateRep), by="5 days"), format="%m-%d", las=2)
axis(2, at=c(10^seq(0,5, 1), 5*10^5), label=c(10^seq(0,5, 1), 5*10^5) )

ymax=sum(filter(covid, countriesAndTerritories==top10Country[1])$deaths)
for (i in 1:10){
    plot(cumsum(deaths)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i]), 
         xlab='', ylab='Total Deaths',ylim=c(1, ymax), log='y',
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n', yaxt='n', cex=1.5)
    lines(cumsum(deaths)~dateRep, filter(covid[order(covid$dateRep),], countriesAndTerritories==top10Country[i]), xlab='',
         col=brewer.pal(n = 10, name = "Spectral")[i], xaxt='n',yaxt='n', lwd=3)
    par(new=TRUE)
}
legend("topleft", legend=top10Country, col=brewer.pal(n = 10, name = "Spectral"),  lty=1, lwd=4, bty='n',y.intersp=0.5)
axis.POSIXct(1, at=seq(min(covid$dateRep), max(covid$dateRep), by="5 days"), format="%m-%d", las=2)
axis(2, at=2*10^seq(0,4, 1), label=2*10^seq(0,4, 1) )

```
